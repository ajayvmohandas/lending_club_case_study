```python
#Features "revol_util" and "int_rate" have "%" symbols, we have to remove that.
loan_data['revol_util'] = loan_data['revol_util'].str.rstrip('%').astype(float)
loan_data['int_rate'] = loan_data['int_rate'].str.rstrip('%').astype(float)
```

```python
loan_data[["revol_util","int_rate"]]
```

```python
loan_data[["issue_d","earliest_cr_line"]]
```

```python
# Columns "issue_d" and "earliest_cr_line" have to be converted to date datatype

loan_data['issue_d'] = pd.to_datetime(loan_data['issue_d'], format='%b-%y')

loan_data['earliest_cr_line'] = pd.to_datetime(loan_data['earliest_cr_line'], format='%b-%y')


# Extract year and month into new columns
loan_data['issue_year'] = loan_data['issue_d'].dt.year
loan_data['issue_month'] = loan_data['issue_d'].dt.month

loan_data['earliest_cr_year'] = loan_data['earliest_cr_line'].dt.year
loan_data['earliest_cr_month'] = loan_data['earliest_cr_line'].dt.month
```

## Handling Outliers

```python
# "annual_inc"
sns.boxplot(loan_data['annual_inc'], orient="h")
```

```python
loan_data['annual_inc'].describe(percentiles=[0.05,0.10,0.25,0.50,0.75,0.9,0.95])
```

```python
#Remove the outliers by considering the threshold as 0.95, because values after 95 percentile seems to be disconnected from the general distribution.
loan_data = loan_data[loan_data['annual_inc'] <= loan_data['annual_inc'].quantile(0.95)]
```
